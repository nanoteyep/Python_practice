```python
def solution(info, query):
    
    class Member_Info:
        def __init__(self, data):
            self.lang = data[0]
            self.job_group = data[1]
            self.career = data[2]
            self.soul_food = data[3]
            self.score = data[4]
            self.left, self.right = None, None
            
    class Member_Search:
        def __init__(self, data):
            self.head = Member_Info(data)
            self.count = 0
            
        def insert(self, data):
            self.current_node = self.head
            while True:
                
                if self.current_node.score <= data[4]:
                    if self.current_node.right == None:
                        self.current_node.right = Member_Info(data)
                        break
                    else:
                        self.current_node = self.current_node.right 
                else:
                    if self.current_node.left == None:
                        self.current_node.left = Member_Info(data)
                        break
                    else:
                        self.current_node = self.current_node.left
                        
        
        def checking(self, target):
                count = 0
                if target[0] == '-' or self.current_node.lang == target[0]:
                    if target[1] == '-' or self.current_node.job_group == target[1]:
                        if target[2] == '-' or self.current_node.career == target[2]:
                            if target[3] == '-' or self.current_node.soul_food == target[3]:
                                count = 1
                return count
            
        def tree_search(self,target):
            
            while self.current_node.right != None or target[4] <= self.current_node.score:
                print(0)
                if target[4] <= self.current_node.score:
                    print(1)
                    self.count += self.checking(target)
                    
                    if self.current_node.left != None and target[4] <= self.current_node.left.score:
                        print(2)
                        self.current_node = self.current_node.left
                        self.tree_search(target)
                    
                    if self.current_node.right != None:
                        print(3)
                        self.curret_node = self.current_node.right
                        self.tree_search(target)
                    
                    break

                else:
                    if self.current_node.right != None and target[4] <= self.current_node.right.score:
                        print(4)
                        self.current_node = self.current_node.right
                        self.tree_search(target)
                    else:
                        break
                        
            return self.count
                    
        def search(self, target):
            self.current_node = self.head
            self.count = 0
            
            self.tree_search(target)
            

                
            return self.count
                
            
#     -------------------------
    
    
    head = info[0]
    head = head.split()
    tree = Member_Search(head)
    
    for data in info[1:]:
        data = data.split()
        tree.insert(data)
        
    answer = []
    
    for target_list in query:
        target_list = target_list.split()
        target = []
        target.append(target_list[0])
        target.append(target_list[2])
        target.append(target_list[4])
        target.append(target_list[6])
        target.append(target_list[7])

        
        count = tree.search(target)
        
        answer.append(count)
        
    return answer
```
